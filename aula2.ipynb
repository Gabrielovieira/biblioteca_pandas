```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP','Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('bandeira').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'].fillna('MS', inplace=True)
vendas.info()
```

```python
media_preco = vendas['Preço'].mean()
vendas['Preço'].fillna(media_preco, inplace=True)
vendas.info()
```

```python
media_preco
```

```python
vendas.head()
```

```python
vendas[['Preço','Preço_com_frete']].describe()
```

```python
vendas_errado = vendas.query('Preço > Preço_com_frete')

vendas_correto = vendas.query('Preço <= Preço_com_frete')
```

```python
vendas_correto.describe()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=True).reset_index(),2)
```

```python
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending=True).reset_index()
```

```python
vendas_correto['mes'] = vendas_correto['Data'].dt.month
vendas_correto.head()
```

```python
vendas_correto.groupby('mes').idcompra.nunique().sort_values(ascending=True).reset_index()
```

```python
cliente = pd.read_excel('cliente_varejo.xlsx')
cliente.head()
```

```python
vendas_correto.head()
```

```python
cliente[['idade', 'renda']].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how= 'left', on= 'cliente_Log')
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').reset_index(),2)
agg_idcanal_renda
```

```python
agg_cliente_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').reset_index(),2)
agg_cliente_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_cliente_bandeira['bandeira'], agg_cliente_bandeira['idade'])
plt.ylabel('Media de idade')
plt.title('Media de idade por bandeira', loc= 'center', fontsize= 18, fontweight= 'bold', color= 'black');
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'])
plt.ylabel('Media de renda')
plt.title('Media de renda por canal de vendas', loc= 'center', fontsize= 18, fontweight= 'bold', color= 'black');
```

```python
venda_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(10,5))

plt.xlabel('Data da venda')
plt.ylabel('Quantidade de vendas')
x = venda_por_data['Data']
y = venda_por_data['idcompra']
plt.plot(x,y, color = 'blue', linewidth = 2)
plt.show()
```

```python
agg_dedt_preco = round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=True).reset_index(),2)
agg_dedt_preco
```

```python
fig = plt.figure(figsize=(30,10))
plt.bar(agg_dedt_preco['Nome_Departamento'],agg_dedt_preco['Preço_com_frete'])
plt.ylabel('Preço medio')
plt.title('Media de preço por departamento', loc= 'center', fontsize= 18, fontweight= 'bold', color= 'black');
```
